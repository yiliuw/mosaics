### Pre-hoc filters

```r
source("./filter.R")
test<-read.csv("/data/sag_data.csv")
options( warn = -1 )
library("xtable")
```

### Cluster level intensity filter

```r
section_filter <- function(data,sec,is_exc){
    li<- output_sample(test, sec)
    if (is_exc == TRUE){    
        type <- li$et
        data<-li$edata
    }
    else{
        type <- li$it
        data<-li$idata
    }
    pp<- create_pp(type,data)
    nn<-calc_R(pp, type, 0.2)
    conf<- calc_conf(pp, type, test, sec)
    res<-intense_qualify(pp,type, nn, conf)
    col.name<-paste('%',toString(sec), sep='')
    df<- data.frame(cluster = res$cluster, coln = round(res$conf,2))
    names(df)[names(df) == "coln"] <- col.name
    return(df)
}
```

### Excitatory clusters    

```r
for (n in 20:24){
    nam <- paste('df',toString(n), sep='')
    assign(nam, section_filter(test, n, TRUE))
}
```

### Inhibitory clusters  
By relaxing on the area requirements, we hope to have more inhibitory cells passing through the filter.

```r
for (n in 20:24){
    nam <- paste('dfi',toString(n), sep='')
    assign(nam, section_filter(test, n, FALSE))
}
```
