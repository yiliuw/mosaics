### L2/3 S24  

```r
subs_cluster<-function(types){
    subs<-c()
    for (t in types){
        sub<- unique(test[test$cluster %in% t,]$subclass)
        subs<-append(subs, sub)
    }
    return(subs)
}
```

```r
df1<- read.csv("/root/capsule/code/table-ex/table-ex.csv")
df1$subs<-subs_cluster(df1$Cluster)
subs.ex<-unique(df1$subs)
```

```r
sec.no<- paste('Zhuang-ABCA-3.0',toString(24), sep='')
sec<-subset(test, brain_section_label ==sec.no)
c1<- sec[sec$subclass %in% '007 L2/3 IT CTX Glut',]
```

```r
ow<-ripras(sec$x, sec$y,shape='convex',f=1)
cm11<- subset(c1, cluster =='0109 L2/3 IT CTX Glut_2')
cm12<- subset(c1, cluster =='0114 L2/3 IT CTX Glut_3')
cm13<- subset(c1, cluster =='0116 L2/3 IT CTX Glut_3')
cm<- c1[!c1$cluster %in% c('0109 L2/3 IT CTX Glut_2','0114 L2/3 IT CTX Glut_3','0116 L2/3 IT CTX Glut_3'),]
```

```r
p1<-affine(ppp(cm11$x, cm11$y,ow),matrix(c(1,0,0,-1),ncol=2))
p2<-affine(ppp(cm12$x, cm12$y,ow),matrix(c(1,0,0,-1),ncol=2))
p3<-affine(ppp(cm13$x, cm13$y,ow),matrix(c(1,0,0,-1),ncol=2))
pm<-affine(ppp(cm$x, cm$y,ow),matrix(c(1,0,0,-1),ncol=2))
```

```r
pp<-create_pp(test, NULL,c(24,23,22,21,20), '0109 L2/3 IT CTX Glut_2')
```

```r
pp1<-create_pp(test, NULL,c(24), '0114 L2/3 IT CTX Glut_3')
```

```r
pp2<-create_pp(test, NULL,c(24,23,21), '0116 L2/3 IT CTX Glut_3')
```

```r
plot(affine(ow,matrix(c(1,0,0,-1),ncol=2)),main='',border="grey", lwd=1)
#plot(Window(pp[[1]]),border='black',lwd=4,add=TRUE)
#plot(Window(pp1[[1]]),border='blue',lwd=4,add=TRUE)
#plot(Window(pp2[[1]]),border='red',lwd=4,add=TRUE)
points(coords(p1)$x, coords(p1)$y, cex=0.5,pch=20, col='red')
points(coords(p2)$x, coords(p2)$y,cex=0.5,pch=20, col='black')
points(coords(p3)$x, coords(p3)$y, cex=0.5,pch=20, col='blue')
points(coords(pm)$x, coords(pm)$y,cex=0.5,pch=20, col='grey')
legend(x = "bottomright", cex=1.2,
       legend=c("0109 L2/3", "0114 L2/3","0116 L2/3","Other L2/3 IT"),  
       fill = c("red","black",'blue','grey')) 
text(6,-3, "Section 24",cex=2, pos=3) 
```
